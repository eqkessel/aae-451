```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as consts
```

```python
V_STALL = np.linspace(0, 50) * consts.foot
RHO = 1.225
CL_MAX = 1.2
B = 5 * consts.foot
M = 10 * consts.lb
AR = np.linspace(0, 10)
```

```python
fig, ax = plt.subplots(1,1)

AR_values = B**2 * (CL_MAX * 0.5 * RHO * (V_STALL)**2) / (M * consts.g)
V_stall_speeds = np.sqrt(2. * M * consts.g * AR / (CL_MAX * B**2))

ax.plot(AR, V_stall_speeds)
ax.plot(AR_values, V_STALL)
ax.grid()
```
